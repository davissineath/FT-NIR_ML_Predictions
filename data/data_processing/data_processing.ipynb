```python
fps = [
    "../glucose_flour_mixes/0/*.csv", 
    "../glucose_flour_mixes/15/*.csv", 
    "../glucose_flour_mixes/25/*.csv",
    "../glucose_flour_mixes/35/*.csv", 
    "../glucose_flour_mixes/45/*.csv", 
    "../glucose_flour_mixes/50/*.csv",
    "../glucose_flour_mixes/55/*.csv",
    "../glucose_flour_mixes/65/*.csv",
    "../glucose_flour_mixes/75/*.csv", 
    "../glucose_flour_mixes/100/*.csv"
    ] 

foods_fps = [
    "../foods/buttermilk_biscuit/*.csv",
    "../foods/inst_potatoes/*.csv",
    "../foods/mixed_gravy/*.csv",
    "../foods/nonfat_milk/*.csv",
    "../foods/parmesan/*.csv",
    "../foods/parmesan_romano/*.csv",
    "../foods/protein_pancake/*.csv",
    "../foods/tapioca/*.csv",
    ]


labels = [0, 0.15, 0.25, 0.35, 0.45, 0.5, 0.55, 0.65, 0.75, 1] 
food_labels = ["buttermilk_biscuit", "inst_potatoes", "mixed_gravy", "nonfat_milk", "parmesan", "parmesan_romano", "protein_pancake", "tapioca"]
```

```python
data = processing_pipeline(fps, labels)
cols_to_drop = [col for col in data.columns if col <= 4500]
data = data.drop(columns=cols_to_drop)

data.to_csv("processed_flour_sugar_data.csv", index=True)
data
```

```python
mean_data= pd.DataFrame(columns=data.columns)
for ind in set(data.index.to_list()):
    mean_data.loc[ind] = data.loc[ind].mean()

mean_data.to_csv("mean_flour_sugar_data.csv", index=True)
```
